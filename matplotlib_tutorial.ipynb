Creating a basic scatter-plot

````python
plt.title("Scatter-plot")
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
plt.scatter(x,a,color="red")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
````

````python
plt.title("Scatter-plot")
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
plt.scatter(x,a,marker="*",c="g",s=100)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
````

Adding two markers in the same plot

````python
plt.title("Scatter-plot")
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]
plt.scatter(x,a,marker="*",c="g",s=500)
plt.scatter(x,b,marker=".",c='r',s=700)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
````

**Adding sub-plots **

````python
plt.title("Scatter-plot")
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]
plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c="g",s=500)
plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c='r',s=700)

plt.show()
````

````python
plt.title("Scatter-plot")
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]
plt.subplot(2,1,1)
plt.scatter(x,a,marker="*",c="g",s=500)
plt.subplot(2,1,2)
plt.scatter(x,b,marker=".",c='r',s=700)

plt.show()
````

Pie -Chart

Creating Data

````python
fruit = ['Apple','Orange','Mango','Gauva']
Quantity = [69,98,78,100]
````

````python
plt.pie(Quantity,labels=fruit)
plt.show()
````

Changing Aesthetics

````python
fruit = ['Apple','Orange','Mango','Gauva']
Quantity = [69,98,78,100]
plt.pie(Quantity,labels=fruit,autopct='%0.2f%%',colors=['yellow','grey','blue','red'])
plt.show()
````

````python
fruit = ['Apple','Orange','Mango','Lichi']
Quantity = [53,68,56,45]
plt.pie(Quantity,labels=fruit,autopct='%0.2f%%',colors=['green','yellow','pink','red'])
plt.show()
````

DoughNut-Chart

````python
fruit = ['Apple','Orange','Mango','Lichi']
Quantity = [67,34,100,29]
plt.pie(Quantity,labels=fruit,radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
````

````python
fruit = ['Apple','Orange','Mango','Lichi']
Quantity = [67,34,100,29]
plt.pie(Quantity,labels=fruit,autopct="%0.1f%%",colors=['green','yellow','orange','pink'],radius=4)
plt.pie([5],colors=['w'],radius=2)
plt.show()
````
